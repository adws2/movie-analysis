```python
from Data_sampling_kobis import crawling_kobis,parsing
import pickle
import os
import numpy as np
import torch
import matplotlib.pyplot as plt
import torch.nn as nn
```

```python
## auto crawling and save xls ##
craw = crawling_kobis()
craw.open()
craw.get_excel_files_all_time(np.arange(0,50))
```

```python
## auto-parsing and save pickle ##
pars = parsing('save')
pars.dict_from_xls()
```

```python
## load file ##
files = []
file_names = os.listdir('{}'.format('save'))
for file in file_names:
    if file[-6:] == 'pickle':
        with open('save/{}'.format(file), 'rb') as fr:
            files.append(pickle.load(fr))
```

```python
## model fitting ##

def mse(y_hat, y):
    return ((y_hat - y) ** 2).mean()


for a_file in files:
    data = []
    for k,v in a_file.items():
        data.append(v)
        
    ## prediction ## 
    
    data = np.array(data)
    beta = torch.tensor(0.25, requires_grad=True)
    #beta = nn.Parameter(beta)
    lr = 0.015

    target = torch.tensor(data,dtype = torch.float)

    target /= data[-1]
    I_0 = target[0]
    x = torch.arange(0,len(target))

    for t in range(500):
        if beta.grad is not None:
            beta.grad.zero_()
        # SI model analytic form
        prediction = I_0 * torch.exp(beta * x) / (1 - I_0 + I_0 * torch.exp(beta * x))
        loss = mse(prediction, target)
        loss.backward()
        beta = (beta.data - lr * beta.grad).requires_grad_(True)
        
    print('{0} , reproductive number : {1}'.format(k[:-3],beta ))
    plt.scatter(x, target,label = 'Actual')
    plt.scatter(x, prediction.detach(),label = 'SI Model')
    plt.legend()
    plt.savefig('save/{}.svg'.format(k[:-3]))
    plt.show()
    
```
